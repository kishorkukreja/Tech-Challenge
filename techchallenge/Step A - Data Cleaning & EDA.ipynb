## Functions & Classes

```python
def get_start_end_dates(year, week):
    d = datetime(year,1,1)
    if(d.weekday()<= 3):
        d = d - timedelta(d.weekday())             
    else:
        d = d + timedelta(7-d.weekday())
    dlt = timedelta(days = (week-1)*7)
    return (d + dlt,  d + dlt + timedelta(days=6))
```
